```python
# ==============================
# Import Pandas Library
# ==============================
import pandas as pd

# ==============================
# Load JSON Data
# ==============================
url = "https://raw.githubusercontent.com/PraveenBunker/45-days-of-python/main/student-data1.json"
df = pd.read_json(url)

# ==============================
# Sort Data by English Marks
# ==============================

# Ascending Order (Default)
df.sort_values("english")

# Descending Order
df = df.sort_values("english", ascending=False)

# Descending Order (Method 2)
df.sort_values(by=["english"], ascending=[False])

# Multiple Column Sorting
# df.sort_values(by=["english", "maths"], ascending=[False, True])

# ==============================
# Sort String Data
# ==============================

# Convert Names to Lowercase
# df["name"] = df["name"].str.lower()

# Sort by Name
# df.sort_values("name")

# ==============================
# Add New Column
# ==============================

df["total"] = df["english"] + df["maths"] + df["physics"]

# ==============================
# Add New Row
# ==============================

df.loc[14] = ["praveen", "male", 50, 80, 80, 210]

# ==============================
# Delete Column
# ==============================

df.drop("total", axis=1)

# ==============================
# Delete Row
# ==============================

# Example:
# df.drop(14, axis=0)

# Display DataFrame
df
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/file2.json"
df = pd.read_json(url)

df['doj'] = ['2025-01-01','2025-02-02','2025-03-03','2025-04-04','2025-05-05']

# Convert to datetime
df['doj'] = pd.to_datetime(df['doj'])

# Date operations
df['year'] = df['doj'].dt.year
df['month'] = df['doj'].dt.month
df['day'] = df['doj'].dt.day
df['day'] = df['doj'].dt.day_name()

# 20 day
df['doj'] = pd.to_timedelta("20 days")


# df['dayofweek'] = df['doj'].dt.dayofweek
# df['dayofyear'] = df['doj'].dt.dayofyear
# df['quarter'] = df['doj'].dt.quarter

print(df)
```
