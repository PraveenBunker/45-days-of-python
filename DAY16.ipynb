## Multiple line chart and legends

```python
import matplotlib.pyplot as plt

x = [2010,2015,2020,2025]
y1 = [100,200,260,290]
y2 = [150,180,215,330]

plt.plot(x,y1,label="jeans")
plt.plot(x,y2,label="T-shirt")
plt.legend()
plt.show()
```

```python
#questio multiple line chart


import matplotlib.pyplot as plt

x = [2000, 2005, 2010, 2015]
y1 = [110, 140, 150, 300]
y2 = [100, 321, 150, 653]
y3 = [200,400,100,50]

plt.plot(x, y1,
         label="jeans",
         color="green",
         marker="o",
         linestyle="--",
         linewidth=4,
         markersize=10)

plt.plot(x, y2,
         label="T-shirt",
         color="pink",
         marker="*",
         linestyle="-",
         linewidth=4,
         markersize=10)

plt.plot(x, y3,
         label="shirt",
         color="red",
         marker="o",
         linestyle=":",
         linewidth=2,
         markersize=10)


plt.legend()
plt.show()
```

## ***Bar chart***

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ["group A", "group B", "group C", "group D"]
y1 = [10, 20, 20, 40]
y2 = [20, 30, 25, 30]

x = np.arange(len(groups))
width = 0.4

plt.bar(x - width/2, y1, width, label="boys", color="red")
plt.bar(x + width/2, y2, width, label="girls", color="blue")

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("No. of students in each group")
plt.xticks(x, groups)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ["group A", "group B", "group C", "group D"]
y1 = [10, 20, 20, 40]
y2 = [20, 30, 25, 30]
y3 = [23, 35, 37, 40]

x = np.arange(len(groups))
width = 0.25

plt.bar(x - width, y1, width, label="boys", color="red")
plt.bar(x, y2, width, label="girls", color="blue")
plt.bar(x + width, y3, width, label="student", color="green")

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("No. of students in each group")
plt.xticks(x, groups)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ["group A", "group B", "group C", "group D"]
y1 = [10, 20, 20, 40]
y2 = [20, 30, 25, 30]
y3 = [23, 35, 37, 40]
y4 = [10, 20, 20, 40]
# y5 = [20, 30, 25, 30]
# y6 = [23, 35, 37, 40]

x = np.arange(len(groups))
width = 0.30

plt.bar(x - width, y1, width, label="boys", color="red")
plt.bar(x , y2, width, label="boys", color="black")
plt.bar(x + width, y3, width, label="girls", color="blue")
plt.bar(x + width, y4, width, label="girls", color="green")

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("No. of students in each group")
plt.xticks(x, groups)
plt.legend()

plt.show()
```

```python
#pie chart

import matplotlib.pyplot as plt

fruits = ["apple","banana", "arange","watermalon"]
count = [40,30,15,70]

plt.pie(count, labels=fruits,autopct="%1.1f%%",colors=['red','green','black','yellow'],startangle=90,explode=(0,0.1,0,0))
# plt.plot(count,fruits)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1st chart
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 55]

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Plot")

# 2nd chart
x1 = ["apple", "banana", "orange", "watermelon"]
y1 = [40, 30, 15, 70]

plt.subplot(2, 2, 2)
plt.pie(y1, labels=x1, startangle=90)
plt.title("Pie Chart")

# 3rd chart
plt.subplot(2, 2, 3)
plt.bar(x1, y1)
plt.title("Bar Chart")

# 4th chart
plt.subplot(2, 2, 4)
plt.scatter(x, y)
plt.title("Scatter Plot")

plt.tight_layout()
plt.show()
```
