```python
#questio multiple line chart


import matplotlib.pyplot as plt

x = [2000, 2005, 2010, 2015]
y1 = [110, 140, 150, 300]
y2 = [100, 321, 150, 653]
y3 = [200,400,100,50]

plt.plot(x, y1,
         label="jeans",
         color="green",
         marker="o",
         linestyle="--",
         linewidth=4,
         markersize=10)

plt.plot(x, y2,
         label="T-shirt",
         color="pink",
         marker="*",
         linestyle="-",
         linewidth=4,
         markersize=10)

plt.plot(x, y3,
         label="shirt",
         color="red",
         marker="o",
         linestyle=":",
         linewidth=2,
         markersize=10)


plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ["group A", "group B", "group C", "group D"]
y1 = [10, 20, 20, 40]
y2 = [20, 30, 25, 30]

x = np.arange(len(groups))
width = 0.4

plt.bar(x - width/2, y1, width, label="boys", color="red")
plt.bar(x + width/2, y2, width, label="girls", color="blue")

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("No. of students in each group")
plt.xticks(x, groups)
plt.legend()

plt.show()
```

# ***Sublots Chart***

```python
#subplots

# one dimension example

import matplotlib.pyplot as plt

year = [2010,2015,2020,2025]
dairy = [100,520,630,400]

#graph 2 data
year1 = [1990,1995,2000,2005]
farming = [300,200,250,100]


fig , aux = plt.subplots(1,2)

aux[0].plot(year,dairy)
aux[0].set_xlabel("year")
aux[0].set_ylabel("dairy")
aux[0].set_title("dairy chart")

aux[1].plot(year1,farming)
aux[1].set_xlabel("year1")
aux[1].set_ylabel("farming")
aux[1].set_title("farming chart")

plt.show()
```

```python
#subplots

# two dimension example

#subplots

# one dimension example

import matplotlib.pyplot as plt

year = [2010,2015,2020,2025]
dairy = [100,520,630,400]

#graph 2 data
year1 = [1990,1995,2000,2005]
farming = [300,200,250,100]


#graph 3 data
year2 = [2012,2015,2020,2025]
dairy2 = [100,150,200,250]

#graph 4 data
year3 = [2010,2015,2020,2025]
farming1 = [500,400,300,200]


fig , aux = plt.subplots(2,2)  #2 row, 2 column

aux[0,0].plot(year,dairy)
aux[0,0].set_xlabel("year")
aux[0,0].set_ylabel("dairy")
aux[0,0].set_title("dairy chart")

aux[0,1].bar(year1,farming)
aux[0,1].set_xlabel("year1")
aux[0,1].set_ylabel("farming")
aux[0,1].set_title("farming chart")


aux[1,0].pie(dairy2, labels=year2)
aux[1,0].set_xlabel("year2")
aux[1,0].set_ylabel("dairy2")
aux[1,0].set_title("dairy2 chart")

aux[1,1].scatter(year3,farming1)
aux[1,1].set_xlabel("year3")
aux[1,1].set_ylabel("farming1")
aux[1,1].set_title("farming1 chart")

plt.tight_layout()


plt.gcf().canvas.get_supported_filetypes()
plt.savefig("subplots.jpg")
plt.show()


```

```python
import matplotlib.pyplot as plt
import pandas as pd

url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-arya/main/temperature_data.json"
df = pd.read_json(url)
# df.drop(2, inplace=True)
df.dropna(subset=["temperature_c"], inplace=True)  #ye column me jitne NaN h usko delete kr dega
print(df)
df["humidity_pct"].mean()

df.fillna(df["humidity_pct"].mean(), inplace=True)
df

df["temperature_f"] = (df["temperature_c"] *1.8) +32
df


#graph -> subplots ->> 1d -> line chart | pie chart

fig, aux = plt.subplots(1,2)

aux[0].plot(df["humidity_pct"], df["temperature_c"])
aux[0].set_xlabel("humidity_pct")
aux[0].set_ylabel("temperature_c")
aux[0].set_title("humidity_pct vs temperature_c")



#second pie graph data

aux[1].pie(df["temperature_f"],labels=df["day"],autopct="%1.1f%%")
aux[1].set_title("temperature_f")

#save image data
plt.savefig("project1.pdf")

plt.show()


```
