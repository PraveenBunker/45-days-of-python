# ***Aggregation***

* Aggregation is used to summarize data.

* It converts multiple values into a single value.

* It helps in analyzing large datasets easily.

* Common aggregation functions are:

sum()

mean()

count()

min()

max()

median()

* The agg() function can apply multiple aggregation functions at once.

* Aggregation is often used with groupby() to analyze grouped data.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/file2.json"
df = pd.read_json(url)
print(df)


#manually
df["marks"].sum()
df["marks"].mean()
df["marks"].max()
df["marks"].min()
df["marks"].std()
df["marks"].median()


#using aggregation
df["marks"].agg(["sum","mean","max","min","std","median"])

```

### ***Concatenate DataFrames***

* Concatenation is used to combine two or more DataFrames.

* It can join DataFrames vertically (rows) or horizontally (columns).

* The pd.concat() function is used for concatenation.

* Useful when datasets have similar structure.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/file2.json"
df = pd.read_json(url)
df1 = pd.read_json(url)

# df1 ke name column me Abhishek ko Praveen karna hai
df1.loc[2,"name"] = "Praveen"

pd.concat([df,df1],axis=1)

```
